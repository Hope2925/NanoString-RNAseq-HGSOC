## Now combine the above to get the full list of 
1. Patients and their clinical data
2. Samples with both Pre and Post NACT, Patient, etc.

```r
# Get the list of all patients and then just combine all possible info
all_patients = union(Laun_only_meta$Patient_code, union(zhang_meta$Patient_ID,
                     union(lahtinen_meta$Patient_ID, pietilla_meta$EOC_ID)))
length(all_patients)
```

```r
patients = c()
Age_list = c() # Zhang Age, 
Data_found_in = c()
OvaHRDscar = c() # Laun
BRCAmut_status = c() # Laun
PFI_12_months = c() # Launonen PFI_category_12_months

CRS = c() # Zhang CRS
CA125_Uperml_TN_PN = c() # Zhang CA125_Uperml_TN and PN

Survival = c() # Pietilla
Residual_tumor_after_DS = c() # Pietilla
Primary_chemo_regimen = c() # Pietilla
Progression_free_surviva_mths = c() # Pietilla
Overall_survival_mths = c() # Pietilla
Relapse_type = c() # Pietilla

PFI_days = c() # Zhang (PFI_days), Pietilla (Platinum_free_interval_days)
FIGO_Stage = c() # Laun Stage_FIGO2014, Zhang Stage, Pietilla FIGO_Stage
Tissues = c()  # Laun is bulkRNAseq, scRNAseq. Zhang is scRNA-seq TN, scRNA-seq PN, Bulk RNA-seq TN Bulk RNA-seq PN. Lahtinen is Tissue. Pietilla is Sample
Sample_stages = c() #Laun is Sample_type, Zhang is NA (all have it I believe), Lahtinen is Treatment phase, Pietilla is Stage_of_sampling
Response_to_Treatment = c() # Lahtinen_meta$treatment_effect (x or not), Pietilla_meta$Primary_therapy_outcome

for (patient in all_patients) {
    patients = c(patients, patient)
    # get patient info if in Launonen (can't be anywhere else)
    if (patient %in% Laun_only_meta$Patient_code) {
        filt = Laun_only_meta[Laun_only_meta$Patient_code == patient,]
        Data_found_in = c(Data_found_in, "Launonen_2024")
        Sample_stages = c(Sample_stages, paste(unique(filt$Sample_type), collapse=","))
        OvaHRDscar = c(OvaHRDscar, filt$OvaHRDscar[1])
        BRCAmut_status = c(BRCAmut_status, filt$BRCAmut_status[1])
        PFI_12_months = c(PFI_12_months, filt$PFI_category_12_months[1])
        FIGO_Stage = c(FIGO_Stage, filt$Stage_FIGO2014[1])
        Tissues = c(Tissues, paste(c(paste(unique(filt$bulkRNAseq), collapse=","), 
                                     paste(unique(filt$scRNAseq), collapse=",")), collapse=","))
        # Info Laun doesn't have
        Age_list = c(Age_list, "NotinData")
        CRS = c(CRS, "NotinData")
        CA125_Uperml_TN_PN = c(CA125_Uperml_TN_PN, "NotinData")
        Response_to_Treatment = c(Response_to_Treatment, "NotinData")
        PFI_days = c(PFI_days, "NotinData")
        Survival = c(Survival, "NotinData")
        Residual_tumor_after_DS = c(Residual_tumor_after_DS, "NotinData")
        Primary_chemo_regimen = c(Primary_chemo_regimen, "NotinData")
        Progression_free_surviva_mths = c(Progression_free_surviva_mths, "NotinData")
        Overall_survival_mths = c(Overall_survival_mths, "NotinData")
        Relapse_type = c(Relapse_type, "NotinData")
    } else {
        # if not in Launonen then
        OvaHRDscar = c(OvaHRDscar, "NotinData")
        BRCAmut_status = c(BRCAmut_status, "NotinData")
        PFI_12_months = c(PFI_12_months, "NotinData")
        Data_found_in_single = c()
        PFI_days_single = c()
        FIGO_Stage_single = c()
        Tissues_single = c()
        Sample_stages_single = c()
        Reponse_to_Treatment_single = c()
        # Consider possibility of multiple samples
        if (patient %in% zhang_meta$Patient_ID) {
            filt = zhang_meta[zhang_meta$Patient_ID == patient,]
            Data_found_in_single = c(Data_found_in_single, "Zhang_2022")
            PFI_days_single = c(PFI_days_single, filt$PFI_days[1])
            FIGO_Stage_single = c(FIGO_Stage_single, paste(unique(filt$Stage[1]), collapse=","))
            Tissues_single = c(Tissues_single, 
                               paste(c(paste(unique(filt[["scRNA-seq TN"]]), collapse=","), 
                                     paste(unique(filt[["scRNA-seq PN"]]), collapse=","), 
                                      paste(unique(filt[["Bulk RNA-seq TN"]]), collapse=","), 
                                     paste(unique(filt[["Bulk RNA-seq PN"]]), collapse=",")), collapse=","))
            Sample_stages_single = c(Sample_stages_single, "TN_PN")
            Age_list = c(Age_list, filt$Age[1])
            # Only Zhang
            CRS = c(CRS, filt$CRS[1])
            CA125_Uperml_TN_PN = c(CA125_Uperml_TN_PN, paste(filt$CA125_Uperml_TN[1], filt$CA125_Uperml_PN[1], collapse=","))
            } else {
            Age_list = c(Age_list, "NotinData")
            CRS = c(CRS, "NotinData")
            CA125_Uperml_TN_PN = c(CA125_Uperml_TN_PN, "NotinData")
        }
        if (patient %in% lahtinen_meta$Patient_ID) {
            filt = lahtinen_meta[lahtinen_meta$Patient_ID == patient,]
            Data_found_in_single = c(Data_found_in_single, "Lahtinen_2023")
            Tissues_single = c(Tissues_single, paste(unique(filt$Tissue), collapse=","))
            Sample_stages_single = c(Sample_stages_single, paste(unique(filt[["Treatment phase"]]), collapse=","))
            Reponse_to_Treatment_single = c(Reponse_to_Treatment_single, filt[["treatment_effect"]][1])
            } 
        # Get Pietilla Specific Info
        if (patient %in%  pietilla_meta$EOC_ID) {
            filt = pietilla_meta[pietilla_meta$EOC_ID == patient,]
            Data_found_in_single = c(Data_found_in_single, "Pietilla_2021")
            PFI_days_single = c(PFI_days_single, filt$Platinum_free_interval_days[1])
            FIGO_Stage_single = c(FIGO_Stage_single, paste(unique(filt$FIGO_stage[1]), collapse=","))
            Tissues_single = c(Tissues_single, paste(unique(filt$Sample), collapse=","))
            Sample_stages_single = c(Sample_stages_single, paste(unique(filt$Stage_of_sampling), collapse=","))
            Reponse_to_Treatment_single = c(Reponse_to_Treatment_single, paste(unique(filt$Primary_therapy_outcome), collapse=","))
            Survival = c(Survival, filt$Survival[1])
            Residual_tumor_after_DS = c(Residual_tumor_after_DS, paste(unique(filt$Residual_tumor_after_DS), collapse=","))
            Primary_chemo_regimen = c(Primary_chemo_regimen, paste(unique(filt$Primary_chemo_regimen), collapse=","))
            Progression_free_surviva_mths = c(Progression_free_surviva_mths, paste(unique(filt$Progression_free_surviva_mths), collapse=","))
            Overall_survival_mths = c(Overall_survival_mths, paste(unique(filt$Overall_survival_mths), collapse=","))
            Relapse_type = c(Relapse_type, paste(unique(filt$Relapse_type), collapse=","))
            } else {
            Survival = c(Survival, "NotinData")
            Residual_tumor_after_DS = c(Residual_tumor_after_DS, "NotinData")
            Primary_chemo_regimen = c(Primary_chemo_regimen, "NotinData")
            Progression_free_surviva_mths = c(Progression_free_surviva_mths, "NotinData")
            Overall_survival_mths = c(Overall_survival_mths, "NotinData")
            Relapse_type = c(Relapse_type, "NotinData")
            }
        # Get the final versions
        Data_found_in = c(Data_found_in, paste(Data_found_in_single, collapse=";"))
        PFI_days = c(PFI_days, paste(PFI_days_single, collapse=";"))
        FIGO_Stage = c(FIGO_Stage, paste(FIGO_Stage_single, collapse=";"))
        Tissues = c(Tissues, paste(Tissues_single, collapse=";"))
        Sample_stages = c(Sample_stages, paste(Sample_stages_single, collapse=";"))
        Response_to_Treatment = c(Response_to_Treatment, paste(Reponse_to_Treatment_single, collapse=";"))
    }
    
    # Check if any of the items are empty?
    }

cat(length(patients), length(Data_found_in), length(PFI_12_months), length(PFI_days), length(FIGO_Stage), 
   length(Tissues), length(Sample_stages), length(Response_to_Treatment), 
   length(Age_list), length(OvaHRDscar), length(Survival))
harmonized_patient_df = data.table("Patient"=patients, "Data_Found_In"=Data_found_in, 
                             "PlatFI_12_months"=PFI_12_months, "PlatFI_days"=PFI_days,
                             "FIGO_Stage"=FIGO_Stage, "Tissues"=Tissues, 
                             "Sample_stages"=Sample_stages, "Response_to_Treatment"=Response_to_Treatment,
                             "Age"=Age_list, "OvaHRDscar"=OvaHRDscar, "BRCAmut_status"=BRCAmut_status, 
                             "CRS"=CRS, "CA125_Uperml_TN_PN"=CA125_Uperml_TN_PN, 
                             "Piet21_Survival"=Survival, "Piet21Residual_tumor_after_DS"=Residual_tumor_after_DS, 
                             "Piet21_Primary_chemo_regimen"=Primary_chemo_regimen, 
                             "Piet21_Progression_free_survival_mths"=Progression_free_surviva_mths, 
                             "Piet21_Overall_survival_mths"=Overall_survival_mths, 
                             "Piet21_Relapse_type"=Relapse_type)
harmonized_patient_df[1:30,]
dim(harmonized_patient_df)
```

```r
# check how many patients actually have both pre and post

pre <- harmonized_patient_df[grepl("chemo-naive|Pre-chemo|TN_PN|Diagnostic", harmonized_patient_df$Sample_stages),]
post <- harmonized_patient_df[grepl("IDS|Post-chemo|TN_PN|Interval", harmonized_patient_df$Sample_stages),]
nrow(pre)
nrow(post)

harmonized_patient_prepost_df <- harmonized_patient_df[harmonized_patient_df$Patient %in% intersect(pre$Patient, post$Patient),]


length(intersect(pre$Patient, post$Patient))
nrow(harmonized_patient_prepost_df)
```

```r
patients = c()
Tissue_list = c()  # Laun is bulkRNAseq, scRNAseq. Zhang is scRNA-seq TN, scRNA-seq PN, Bulk RNA-seq TN Bulk RNA-seq PN. Lahtinen is Tissue. Pietilla is Sample
RNAType_list = c()
Data_found_in = c()
RNAseqID = c()
Sample_stages = c() #Laun is Sample_type, Zhang is NA (all have it I believe), Lahtinen is Treatment phase, Pietilla is Stage_of_sampling


for (patient in all_patients) {
    # get patient info if in Launonen (can't be anywhere else)
    if (patient %in% Laun_only_meta$Patient_code) {
        filt = Laun_only_meta[Laun_only_meta$Patient_code == patient,]
        for (index in seq(1, nrow(filt))) {
             Data_found_in = c(Data_found_in, "Launonen_2024")
             Sample_stages = c(Sample_stages, filt$Sample_type[index])
             RNAType_list = c(RNAType_list, filt$RNA_status[index])
             RNAseqID = c(RNAseqID, NA)
             Tissue_list = c(Tissue_list, filt$bulkRNAseq[index])
             patients = c(patients, patient)
            } }
    if (patient %in% zhang_meta$Patient_ID) {
        filt = zhang_meta[zhang_meta$Patient_ID == patient,]
        Data_found_in = c(Data_found_in, "Zhang_2022", "Zhang_2022")
        Sample_stages = c(Sample_stages, "Treatment-Naive", "Post-NACT")
        #### RNAseqID = c(RNAseqID, NA)
        patients = c(patients, patient, patient)
        RNAseqID = c(RNAseqID, NA, NA)
        for (index in seq(1, nrow(filt))) {
            if (is.na(filt[["scRNA-seq TN"]][index])) {
            Tissue_list = c(Tissue_list, filt[["scRNA-seq TN"]][index], filt[["scRNA-seq PN"]][index])
            RNAType_list = c(RNAType_list, "SC", "SC")
            } else {
            Tissue_list = c(Tissue_list, filt[["Bulk RNA-seq TN"]][index], filt[["Bulk RNA-seq PN"]][index])
            RNAType_list = c(RNAType_list, "Bulk", "Bulk")
            }}
        }
             
     if (patient %in% lahtinen_meta$Patient_ID) {
        filt = lahtinen_meta[lahtinen_meta$Patient_ID == patient,]
        for (index in seq(1, nrow(filt))) {
            patients = c(patients, patient)
            Data_found_in = c(Data_found_in, "Lahtinen_2023")
            RNAType_list = c(RNAType_list, "Bulk")
            Tissue_list = c(Tissue_list, filt$Tissue[index])
            RNAseqID = c(RNAseqID, filt$Sample_IDa[index])
            Sample_stages = c(Sample_stages, filt[["Treatment phase"]][index])
            }}
             
     if (patient %in% pietilla_meta$EOC_ID) {
        filt = pietilla_meta[pietilla_meta$EOC_ID == patient,]
        for (index in seq(1, nrow(filt))) {
            patients = c(patients, patient)
            RNAseqID = c(RNAseqID, filt$RNASeq_ID[index])
            Data_found_in = c(Data_found_in, "Pietilla_2021")
            RNAType_list = c(RNAType_list, "Bulk")
            Tissue_list = c(Tissue_list, filt$Sample[index])
            Sample_stages = c(Sample_stages, filt[["Stage_of_sampling"]][index])
            }}
}


cat(length(patients), length(Tissue_list), length(RNAType_list), length(Data_found_in), length(RNAseqID), 
   length(Sample_stages))
harmonized_sample_df = data.table("Patient"=patients, "Data_Found_In"=Data_found_in, 
                             "Assay"=RNAType_list, "Tissue"=Tissue_list, 
                             "RNAseqID"=RNAseqID, "Stage"=Sample_stages)

unique(harmonized_sample_df$Stage)
# get harmonized names for stage
harmonized_sample_df[harmonized_sample_df$Stage %in% c("chemo-naive", "Treatment-Naive", 
                                                      "Diagnostic", "Pre-chemo"),]$Stage <- "PreNACT"
harmonized_sample_df[harmonized_sample_df$Stage %in% c("Post-NACT", "Interval", 
                                                      "IDS", "Post-chemo"),]$Stage <- "PostNACT"
#harmonized_sample_df
harmonized_sample_df$Comb_Name_Tissue <- paste0(harmonized_sample_df$Patient, "_", harmonized_sample_df$Tissue, "_",
                                        harmonized_sample_df$Stage, "_",harmonized_sample_df$Assay)
harmonized_sample_df$Comb_Name <- paste0(harmonized_sample_df$Patient, "_", 
                                        harmonized_sample_df$Stage, "_",harmonized_sample_df$Assay)
#table(harmonized_sample_df$Comb_Name)
dim(harmonized_sample_df)


# How many samples have a twin (Same everything except Pre vs Post)
Twin_list = c()
Twin_bool_list = c()
Exact_Twin_list = c()
Exact_Twin_bool_list = c()

for (index in seq(1, nrow(harmonized_sample_df))) {
    name = harmonized_sample_df$Comb_Name[index]
    exact_name = harmonized_sample_df$Comb_Name_Tissue[index]
    stage = harmonized_sample_df$Stage[index]
    if (is.na(stage)) {
        Twin_list = c(Twin_list, NA)
        Twin_bool_list = c(Twin_bool_list, NA)
        Exact_Twin_list = c(Exact_Twin_list, NA)
        Exact_Twin_bool_list = c(Exact_Twin_bool_list, NA)
        } else if (stage == "PreNACT") {
         twin = harmonized_sample_df[harmonized_sample_df$Comb_Name == sub("Pre", "Post", name),]
         if (nrow(twin) > 0) {
             Twin_list = c(Twin_list, twin$Comb_Name[1])
             Twin_bool_list = c(Twin_bool_list, "Twin")
             } else {
             Twin_list = c(Twin_list, "None")
             Twin_bool_list = c(Twin_bool_list, "None")
             }
         twin = harmonized_sample_df[harmonized_sample_df$Comb_Name_Tissue == sub("Pre", "Post", exact_name),]
         if (nrow(twin) > 0) {
             Exact_Twin_list = c(Exact_Twin_list, twin$Comb_Name_Tissue[1])
             Exact_Twin_bool_list = c(Exact_Twin_bool_list, "Exact_Twin")
             } else {
             Exact_Twin_list = c(Exact_Twin_list, "None")
             Exact_Twin_bool_list = c(Exact_Twin_bool_list, "None")
             }
        } else if (stage == "PostNACT") {
        twin = harmonized_sample_df[harmonized_sample_df$Comb_Name == sub("Post", "Pre", name),]
        if (nrow(twin) > 0) {
              Twin_list = c(Twin_list, twin$Comb_Name[1])
            Twin_bool_list = c(Twin_bool_list, "Twin")
             } else {
             Twin_list = c(Twin_list, "None")
            Twin_bool_list = c(Twin_bool_list, "None")
             }
         twin = harmonized_sample_df[harmonized_sample_df$Comb_Name_Tissue == sub("Post", "Pre", exact_name),]
         if (nrow(twin) > 0) {
             Exact_Twin_list = c(Exact_Twin_list, twin$Comb_Name_Tissue[1])
             Exact_Twin_bool_list = c(Exact_Twin_bool_list, "Exact_Twin")
             } else {
             Exact_Twin_list = c(Exact_Twin_list, "None")
             Exact_Twin_bool_list = c(Exact_Twin_bool_list, "None")
             }
        } else {
        Twin_list = c(Twin_list, NA)
        Twin_bool_list = c(Twin_bool_list, NA)
        Exact_Twin_list = c(Exact_Twin_list, NA)
        Exact_Twin_bool_list = c(Exact_Twin_bool_list, NA)
        }
    
    
    }
length(Twin_list)
sum
harmonized_sample_df$Twin = Twin_list
harmonized_sample_df$Twin_bool = Twin_bool_list
harmonized_sample_df$Exact_Twin = Exact_Twin_list
harmonized_sample_df$Exact_Twin_bool = Exact_Twin_bool_list

harmonized_sample_df[1:4,]
table(harmonized_sample_df$Twin_bool)
table(harmonized_sample_df$Exact_Twin_bool)

length(unique(harmonized_sample_df[harmonized_sample_df$Twin_bool == "Twin",]$Patient))
length(unique(harmonized_sample_df[harmonized_sample_df$Exact_Twin_bool == "Exact_Twin",]$Patient))

harmonized_sample_twin_df = harmonized_sample_df[harmonized_sample_df$Twin_bool == "Twin",]
```
